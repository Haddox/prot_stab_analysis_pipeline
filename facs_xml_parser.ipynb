```python
import xml.etree.ElementTree as ET

import os
from glob import glob
```

```python
xml_dir = "/home/jupyter/sd2e-community/shared-q1-workshop/strcklnd/facs_data/production_data/sort_reports/"

for sub_dir in glob(xml_dir + '*/'):
    for filename in os.listdir(sub_dir):
        tree = ET.parse(os.path.join(sub_dir, filename))
        root = tree.getroot()

        exp = root.findall(".//*[@name='Experiment']")[0].get('value')
        spec = root.findall(".//*[@name='Specimen']")[0].get('value')
        tube = root.findall(".//*[@name='Tube']")[0].get('value')
        
        counts = root[4].findall(".//*[@col='1']")[0].text
        
        print(exp + " | " + spec + " | " + tube + " | " + str(counts) + "\n")
```
